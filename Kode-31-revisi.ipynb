```python
import os
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration

print("Semua pustaka berhasil diimpor!")
```

```python
# Inisialisasi model BLIP
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

def generate_caption(image_path):
    """Menghasilkan deskripsi gambar menggunakan BLIP."""
    try:
        image = Image.open(image_path).convert("RGB")
        inputs = processor(image, return_tensors="pt")
        with torch.no_grad():
            output = model.generate(**inputs)
        caption = processor.decode(output[0], skip_special_tokens=True)
        return caption
    except Exception as e:
        print(f"Error dalam memproses {image_path}: {e}")
        return None

def process_images_in_folder(root_folder, output_file):
    """Memproses semua gambar dalam folder dan subfolder, lalu menyimpan hasil ke CSV."""
    try:
        with open(output_file, "w", encoding="utf-8") as file:
            file.write("nama_file,caption\n")  # Header kolom
            
            for folder_path, _, filenames in os.walk(root_folder):
                for img_name in filenames:
                    img_path = os.path.join(folder_path, img_name)
                    
                    if img_name.lower().endswith((".png", ".jpg", ".jpeg")):
                        caption = generate_caption(img_path)
                        if caption:
                            filename = os.path.basename(img_path)
                            caption_clean = caption.replace('"', "'").strip()
                            file.write(f"{filename},\"{caption_clean}\"\n")
                            print(f"✅ {filename}: {caption_clean}")
                        else:
                            print(f"⚠️ Gagal memproses: {img_name}")
        
        print("\n✅ Semua gambar telah berhasil diproses dan disimpan ke:", output_file)
    except Exception as e:
        print(f"❌ Terjadi kesalahan besar: {e}")

# Jalankan
folder = "dataset-31-v2"
output_file = "captions-31-v2.csv"
process_images_in_folder(folder, output_file)
```
